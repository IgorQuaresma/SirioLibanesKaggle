#Exploratory Analysis

## Introduction

This notebook is the part 2 of 3 other notebooks that composes the project. The main purpose of this notebook is to do an exploratory analysis of the data. This notebook uses the "processed_data" dataframe from part 1. 



## Importing the libs

```python
#Import libs
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Functions

```python
def count_plot_ICU(data, title='', xlabel='', ylabel='', cat0='', cat1=''):
  '''Create a count plot of the amount of cases that needed treatment in the ICU
  and amount of cases that didn`t need ICU treatment'''

  #Creating chart
  sns.set_style('white')
  fig, ax = plt.subplots(figsize=(10,6))
  sns.countplot(x='ICU', data=data,palette=['royalblue','lightcoral'],alpha = 0.6)

  #Creating chart elements
  plt.title(title, fontsize=18)
  plt.xlabel(xlabel, fontdict={'fontsize': 14})
  plt.ylabel(ylabel, {'fontsize': 14})
  plt.ylim((0,250))
  sns.despine()

  #Inserting values on top of each bar
  total = []
  for p in ax.patches:
    total.append(p.get_height())
  total = sum(total)

  for p in ax.patches:
    ax.text(p.get_x()+.30, p.get_height()+1.9, \
    str(round((p.get_height()/total)*100, 2))+'%',
    fontsize=15)

  labels = [item.get_text() for item in ax.get_xticklabels()]
  labels[0] = cat0
  labels[1] = cat1

  ax.set_xticklabels(labels)

  for i in ax.get_yticklabels() + ax.get_xticklabels():
    i.set_fontsize(14)
  for i in ax.get_yticklabels() + ax.get_xticklabels():
    i.set_fontsize(14)
  plt.show()
```

```python
def count_plot_gender(data, title='', xlabel='', ylabel='', cat0='', cat1=''):
  '''Create a count plot of the distribution of gender in the dataset'''

  #Creating chart
  sns.set_style('white')
  fig, ax = plt.subplots(figsize=(10,6))
  sns.countplot(x='GENDER', data=data, palette=['royalblue','lightcoral'], alpha = 0.6)

  #Creating main chart elements
  plt.title(title, fontsize=18)
  plt.xlabel(xlabel, fontdict={'fontsize': 14})
  plt.ylabel(ylabel, {'fontsize': 14})
  plt.ylim((0,250))
  sns.despine()

  #Inserting values on top of each bar
  total = []
  for p in ax.patches:
    total.append(p.get_height())
  total = sum(total)

  for p in ax.patches:
    ax.text(p.get_x()+.30, p.get_height()+1.9, \
    str(round((p.get_height()/total)*100, 2))+'%',
    fontsize=15)

  labels = [item.get_text() for item in ax.get_xticklabels()]
  labels[0] = cat0
  labels[1] = cat1

  ax.set_xticklabels(labels)

  for i in ax.get_yticklabels() + ax.get_xticklabels():
    i.set_fontsize(14)
  for i in ax.get_yticklabels() + ax.get_xticklabels():
    i.set_fontsize(14)
  plt.show()
```

```python
def count_plot_age_percentil(data, title='', xlabel='', ylabel='', cat0='', cat1=''):
  '''Create a count plot of the distribution of gender in the dataset'''

  #Creating chart
  sns.color_palette("rocket")
  fig, ax = plt.subplots(figsize=(20,8))
  sns.countplot(x='AGE_PERCENTIL', data=data, alpha = 0.6, palette='rocket', hue='ICU')

  #Creating main chart elements
  plt.title(title, fontsize=25)
  plt.xlabel(xlabel, fontdict={'fontsize': 14})
  plt.xticks(rotation=45)
  plt.ylabel(ylabel, {'fontsize': 14})
  plt.ylim((0,40))
  plt.legend(['doesn`t need ICU support','needs ICU support'], fontsize=14 )
  sns.despine()

  #Inserting values on top of each bar
  total = []
  for p in ax.patches:
    total.append(p.get_height())
  total = sum(total)

  for p in ax.patches:
    ax.text(p.get_x()+.01, p.get_height()+1.15, \
    str(round((p.get_height()/total)*100, 2))+'%',
    fontsize=14)

  #Adjusting ticks 
  labels = [item.get_text() for item in ax.get_xticklabels()]
  labels[0] = '10th'
  labels[1] = '20th'
  labels[2] = '30th'
  labels[3] = '40th'
  labels[4] = '50th'
  labels[5] = '60th'
  labels[6] = '70th'
  labels[7] = '80th'
  labels[8] = '90th'
  labels[9] = 'ABOVE 90th'

  ax.set_xticklabels(labels)

  for i in ax.get_yticklabels() + ax.get_xticklabels():
    i.set_fontsize(14)
  for i in ax.get_yticklabels() + ax.get_xticklabels():
    i.set_fontsize(14)
  plt.show()
```

```python
def count_plot_gender_ICU(data, title='', xlabel='', ylabel='', cat0='', cat1=''):
  '''Create a count plot of the distribution of gender and necessity of  ICU support'''

  #Creating chart
  sns.set_style('white')
  fig, ax = plt.subplots(figsize=(10,6))
  sns.countplot(x='ICU', hue='GENDER' ,data=data, palette=['royalblue','lightcoral'], alpha = 0.6)

  #Creating main chart elements
  plt.title(title, fontsize=18)
  plt.xlabel(xlabel, fontdict={'fontsize': 14})
  plt.ylabel(ylabel, {'fontsize': 14})
  plt.ylim((0,250))
  plt.legend(['Men','Women'], fontsize=14 )
  sns.despine()

  #Inserting values on top of each bar
  total = []
  for p in ax.patches:
    total.append(p.get_height())
  total = sum(total)

  for p in ax.patches:
    ax.text(p.get_x()+.12, p.get_height()+1.9, \
    str(round((p.get_height()/total)*100, 2))+'%',
    fontsize=15)

  labels = [item.get_text() for item in ax.get_xticklabels()]
  labels[0] = cat0
  labels[1] = cat1

  ax.set_xticklabels(labels)

  for i in ax.get_yticklabels() + ax.get_xticklabels():
    i.set_fontsize(14)
  for i in ax.get_yticklabels() + ax.get_xticklabels():
    i.set_fontsize(14)
  plt.show()
```

```python
def count_plot_disease(data, title='', xlabel='', ylabel='', cat0='', cat1=''):
  '''Create a count plot of which diseases group were the ones that most appeared in patients that went to the ICU'''

  #Creating melt dataframe
  diseases = data[['DISEASE GROUPING 1','DISEASE GROUPING 2','DISEASE GROUPING 3','DISEASE GROUPING 4','DISEASE GROUPING 5','DISEASE GROUPING 6','HTN','OTHER','ICU']]
  melted_diseases_data  = pd.melt(diseases, id_vars='ICU')
  melted_diseases_data

  #Creating chart
  sns.color_palette("rocket")
  fig, ax = plt.subplots(figsize=(15,10))
  sns.countplot(x='variable', data=melted_diseases_data.query('value == 1'), hue='ICU', alpha = 0.6, palette='rocket')

  #Creating main chart elements
  plt.title(title, fontsize=25)
  plt.xlabel(xlabel, fontdict={'fontsize': 14})
  plt.xticks(rotation=45)
  plt.ylabel(ylabel, {'fontsize': 14})
  plt.ylim((0,200))
  plt.legend(['doesn`t need ICU support','needs ICU support'], fontsize=14 )
  sns.despine()

  #Inserting values on top of each bar
  total = []
  for p in ax.patches:
    total.append(p.get_height())
  total = sum(total)

  for p in ax.patches:
    ax.text(p.get_x()+.01, p.get_height()+1.2, \
    str(round((p.get_height()/total)*100, 2))+'%',
    fontsize=13)

  #Adjusting ticks 
  labels = [item.get_text() for item in ax.get_xticklabels()]

  ax.set_xticklabels(labels)

  for i in ax.get_yticklabels() + ax.get_xticklabels():
    i.set_fontsize(12)
  for i in ax.get_yticklabels() + ax.get_xticklabels():
    i.set_fontsize(12)
  plt.show()
```

###Sírio Libanês data analysis

This part of the project will cover the exploratory analysis of the Sírio Libanês data. It is going to be used the dataframe "processed_data" from the part 1 of the project as the input data of this notebook.

As a way to organize the exploratory analysis, the following questions will be answered:


1.   How is the gender distributed in our dataset? What is the number of men and women in our data?
2.  How many patients needed to go to the ICU for treatment?
3.  What is the distribution of patients that needed ICU support in each age percentil?
4.  Which gender is more predominant in the necessity of using the ICU unit?
5.  Which disease group appears to be the responsible for most of the ICU admissions in our dataset? 


```python
data = pd.read_csv("https://raw.githubusercontent.com/IgorQuaresma/SirioLibanesKaggle/main/Data/Processed/processed_data_ea.csv")
data.head()
```

**1. How is the gender distributed in our dataset? What is the number of men and women in our data?**

```python
count_plot_gender(data, title='Distribution of patients in each gender', xlabel='Gender', ylabel='Count', cat0='Men', cat1='Women')
```

From the total of **352 patients**, there are **218 men** and **134 women**.

**2. How many patients needed to go to the ICU for treatment?**

```python
count_plot_ICU(data, title='Necessity to use ICU', xlabel='Support Use', ylabel='Count', cat0='doesn`t need ICU support', cat1='needs ICU support')
```

The total number of **163 patients (46.31%)** needed the ICU unit support according to our dataset

**3. What is the distribution of patients that needed ICU support in each age percentil?**

```python
count_plot_age_percentil(data, title='Distribution of patients per age percentil and gender', xlabel='Age Percentil', ylabel='Count')
```

Using the chart above, it is possible to see that  **the need to use the ICU support increases with the age of the patients**. That supports why older people were the most affected by the Covid-19, especially, when we are considering the death rate. At the beginning of the Covid-19 pandemic, the hospitals were not prepared for the huge number of patients that needed ICU support. This way, **older people that needed a more specialized treatment did not have this opportunity**, resulting in the death of many of those seniors.

In contrast of what happens to eldery people, **the 10th, 20th, 30th, 40th, 50th, and 60th age percentils have less patients that needed ICU support**. It is possible to conclude that younger people tended to go less to the ICUs for treatment.

**4. Which gender is more predominant in the necessity of using the ICU unit?**

```python
count_plot_gender_ICU(data, title='ICU support by each gender', xlabel='Support Use', ylabel='Count', cat0='doesn`t need ICU support', cat1='needs ICU support')
```

**The masculine gender was the one that needed more ICU support (31.53%)** This is twice as much than the number of women that needed the same treatment. It is important to remember that the masculine gender represents more than 50% of the data. Taking this into consideration, **more than 50% of the whole masculine gender patients in our dataset went to the ICU, against  38% from the feminine gender.**

**5. Which disease group appears to be the responsible for most of the ICU admissions in our dataset?**

```python
count_plot_disease(data, title='Which disease group appeared the most in ICU patients', xlabel='', ylabel='Count')
```

**The "other" cause was the one that most appeared for patients that went and did not go to the ICU**. THis is mainly because all other reasons that are not included in HTN and diseases groups 1 to 6 are included in this category. Besides that, the HTN was the second category that predominated the most in patients analyzed in this project.

As we can see, **diseases groups 2, 4 and 5 were the ones that least appeared in patients that went to the ICU**.

## Conclusion

The ICU admission occured the most with:
